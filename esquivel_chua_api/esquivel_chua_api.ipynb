```python
from newsapi import NewsApiClient
import pandas as pd
import json

api = NewsApiClient('7de1dddfaeab4093ba7df0e5574c6bdd')
```

```python
# Get list of articles from June 1 and 2
news_list = api.get_everything(domains='inquirer.net', from_param='2022-06-01', to='2022-06-02')
news_list
```

```python
# Write to json
with open('esquivel_chua_api.json', 'w') as fp:
    json.dump(news_list['articles'], fp, indent = 4)
```
